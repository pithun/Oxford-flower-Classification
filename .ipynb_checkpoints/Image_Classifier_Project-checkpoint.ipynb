```python
# Import TensorFlow 
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_hub as hub

# Ignore some warnings that are not relevant 
import warnings
warnings.filterwarnings('ignore')

# All other necessary imports.
import matplotlib.pyplot as plt
import json
import numpy as np
import os

# Some other visualization settings
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
tfds.disable_progress_bar()
```

## Loading the Dataset

```python
data, info = tfds.load('oxford_flowers102', as_supervised = True, with_info = True)

# Creating a training set, a validation set and a test set.
training_set = data['train']
val_set = data['validation']
test_set = data['test']
```

## Exploring the Dataset

```python
# Getting the number of examples in each set from the dataset info.
for a in ['train', 'validation', 'test']:
    print('The number of images in the', a, 'set is', info.splits[a].num_examples)


# Getting the number of classes in the dataset from the dataset info.
print('\n' + 'There are', info.features['label'].num_classes, 'features')
```

```python
# Printing the shape and corresponding label of 3 images in the training set.
for img, lab in training_set.take(3):
    print('Image shape is:', img.shape)
    print('Image label is:', lab.numpy())
```

```python
# Plotting 1 image from the training set. 
for img, lab in training_set.take(1):
    image = img.numpy().squeeze()
    label = lab.numpy()

plt.imshow(image, cmap = plt.cm.binary)
plt.colorbar()

# Setting the title of the plot to the corresponding image label. 
plt.title(label)
plt.show()
```

### Label Mapping
Dictionary mapping the integer coded labels to the actual names of the flowers.

```python
with open('label_map.json', 'r') as f:
    class_names = json.load(f)
```

```python
# Plotting 1 image from the training set and setting the title of the plot to the corresponding class name. 
for img, lab in training_set.take(1):
    image = img.numpy().squeeze()
    label = lab.numpy()

plt.imshow(image, cmap = plt.cm.binary)
plt.colorbar()

# Setting the title of the plot to the corresponding image label. 
plt.title(class_names[str(label+1)])
plt.show()
```

## Creating Pipeline

```python
# Creating a pipeline for each set.
batch_size = 30
image_size = 224

def normalize_and_resize(img, lab):
    img = tf.cast(img, tf.float32)
    img = tf.image.resize(img, (image_size, image_size))
    img /= 255
    return img, lab

tr_batches = training_set.shuffle(30000//4).map(normalize_and_resize).batch(batch_size).prefetch(1)
val_batches = val_set.map(normalize_and_resize).batch(batch_size).prefetch(1)
test_batches = test_set.map(normalize_and_resize).batch(batch_size).prefetch(1)
```

```python
# verifying sizes
for img_batch, lab in tr_batches.take(1):
    print(img_batch.shape)
    for img, label in zip(img_batch, lab):
        image = img.numpy().squeeze()
        label = label.numpy()
        break

plt.imshow(image, cmap = plt.cm.binary)
plt.colorbar()

# Setting the title of the plot to the corresponding image label. 
plt.title(class_names[str(label+1)].title())
plt.show()

# It's confirmed that the images have been resized and the pixels normalized.
```

# Building and Training the Classifier using Transfer Learning

```python
# Importing model from tenssorflow hub
url = 'https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4'
feature_extract = hub.KerasLayer(url, input_shape = (image_size, image_size, 3))
feature_extract.trainable = False

model = tf.keras.Sequential([
    feature_extract,
    tf.keras.layers.Dense(102, activation = 'softmax')
])

model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
# Early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss', patience = 5)

#saving best model
save_best = tf.keras.callbacks.ModelCheckpoint("./best_model.h5", monitor = 'val_loss',
                                              save_best_only = True)
```

```python
# testing_untrained model
for batch, lab in tr_batches.take(1):
    loss, accuracy = model.evaluate(batch, lab)
print('loss =', loss)
print('accuracy =', accuracy )
```

```python
# fiting the model
epochs = 50
history = model.fit(tr_batches,
                       epochs = epochs,
                       validation_data = val_batches,
                       callbacks = [early_stopping, save_best])
```

```python
# Plotting the loss and accuracy values achieved during training for the training and validation set.
train_accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']

train_loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(len(train_accuracy))

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.plot(epochs_range, train_accuracy, label = 'train_accuracy')
plt.plot(epochs_range, val_accuracy, label = 'val_accuracy')
plt.legend()
plt.title('Training and Val accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range, train_loss, label = 'train_loss')
plt.plot(epochs_range, val_loss, label = 'val_loss')
plt.legend()
plt.title('Training and Val loss')
```

## Testing the Network

```python
# Printing the loss and accuracy values achieved on the entire test set.
loss, accuracy = model.evaluate(test_batches)
print(loss, accuracy)
```

# Inference for Classification


```python
# Creating the process_image function that normalizes the pixels
def process_image(img):
    img = tf.cast(img, tf.float32)
    img = tf.image.resize(img, (image_size, image_size))
    img /= 255
    return img
```

```python
# Testing the process image function by loading one of the above images using PIL and plotting the original image alongside the 
# image produced by the process_image function. If the process_image function works, the plotted image should be the 
# correct size.

from PIL import Image

image_path = './water_lily.jpg'
im = Image.open(image_path)
test_image = np.asarray(im)

processed_test_image = process_image(test_image)

fig, (ax1, ax2) = plt.subplots(figsize=(10,10), ncols=2)
ax1.imshow(test_image)
ax1.set_title('Original Image')
ax2.imshow(processed_test_image)
ax2.set_title('Processed Image')
plt.tight_layout()
plt.show()
```

```python
# Creating the predict function which processes an image and then uses the trained model to make a prediction.

def predict(image_path, model, top_k, plot_image = True, prob_chart = True):
    image_path = image_path
    im = Image.open(image_path)
    test_image = np.asarray(im)
    vals = model.predict(np.expand_dims(process_image(test_image), axis = 0))
    
    values, ind = tf.nn.top_k(vals, k = top_k)
    
    # conerting to list and storing values
    probs = list(values.numpy()[0])
    classes = list(ind.numpy()[0])
    
    # Plotting image
    if plot_image == True and prob_chart == True:
        plt.figure(figsize = (20,5))
        
        # getting the image name
        # after spliting by '/' the value with the max length should be the flowername
        path_split = image_path.split('/')
        im = [len(c) for c in path_split]
        im_name = path_split[im.index(max(im))]
        plt.subplot(1,2,1)
        plt.imshow(process_image(test_image))
        plt.title(im_name[:-4])
        
        plt.subplot(1,2,2)
        plt.barh([class_names[str(a+1).title()] for a in classes], probs)
        plt.title('Class Probability')
        plt.show()
        
    elif plot_image == False and prob_chart == False:
        pass 
    
    elif plot_image == False and prob_chart == True:
        plt.barh([class_names[str(a)] for a in classes], probs)
        plt.title('Class Probability')
        plt.show()
        
    elif plot_image == True and prob_chart == False:
        plt.imshow(process_image(test_image))
        plt.show()
    
    else:
        raise ValueError("Enter either True or False to the \"plot_image\" arg")
    
    return probs, classes
```
